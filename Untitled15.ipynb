```python
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
import nltk
from nltk.stem import PorterStemmer
from nltk.corpus import state_union
from nltk.tokenize import PunktSentenceTokenizer
from nltk.corpus import wordnet
import random
```

```python
short_pos = open('positive.txt','r').read()
short_neg  =open('negative.txt','r').read()
```

```python
documents = []
for r in short_pos.split('\n'):
    documents.append((r,'pos'))
for r in short_neg.split('\n'):
    documents.append((r,'neg'))    
```

```python
all_words = []
short_neg_words = word_tokenize(short_neg)
short_pos_words = word_tokenize(short_pos)

for w in short_neg_words :
    all_words.append(w.lower())
for w in short_pos_words:
    all_words.append(w.lower())    
```

```python
all_words = nltk.FreqDist(all_words)
word_features = list(all_words.keys())[:5000]
def find_features(document):
    words = word_tokenize(document)
    features = {}
    for w in word_features:
        features[w] = (w in words)
    return features    
```

```python
featuresets = [(find_features(rev),category) for (rev,category) in documents]
random.shuffle(featuresets)
```

```python
training_data = featuresets[:10000]
test_data = featuresets[10000:]
```

```python
from nltk.classify.scikitlearn import SklearnClassifier
```

```python
from sklearn.naive_bayes import MultinomialNB,GaussianNB,BernoulliNB
```

```python
classifier = nltk.NaiveBayesClassifier.train(training_data)
```

```python
print(nltk.classify.accuracy(classifier,test_data)*100)
```

```python
def sentiment(text):
    feats = find_features(text)
    return classifier.classify(feats),feats
```

```python
print(sentiment('This movie was awesome! The acting was great,plot was wonderful, and there were pythons....so yea!'))
```
